### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:  

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
import numpy as np
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
# Создаю фильтр за счет метода loc. 
# Каждое условие прописываю в отдельный фильтр отдельной строкой
# Почему не обычными приемами фильтрации? 
# Потому что метод loc может создавать и заполнять столбци самостоятельно без написания доп функций с проверками и условими


# 1 если источник traffic_source равен yandex или google, то в source_type ставится organic
log.loc[(log['traffic_source'] == 'yandex') | 
        log['traffic_source'] == 'google' , 'source_trafic'] = 'organic'

# - Условие читается так: если в столбце log['traffic_source'] есть значение 'yandex' и 'google', 
# - создай столбец 'source_trafic' и заполни его 'organic' ...дале все цсловия по аналогии

# 2 для источников paid и email из России - ставим ad
log.loc[((log['traffic_source'] == 'paid') | 
         (log['traffic_source'] == 'email')) & 
        (log['region'] == 'Russia') , 'source_trafic'] = 'ad'
# 3 для источников paid и email не из России - ставим other
log.loc[((log['traffic_source'] == 'paid') | 
         (log['traffic_source'] == 'email')) 
        != (log['region'] == 'Russia') , 'source_trafic'] = 'other'
# 4 все остальные варианты берем из traffic_source без изменений
log.loc[log['traffic_source'] == log ['traffic_source'], 'source_trafic'] = log['traffic_source']
log.head(10)
```

### Задание 3

- Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
- Посчитайте **среднее время жизни пользователей**, которые **выставили более 100 оценок**. 
<!-- - Под **временем жизни** понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId. -->

```python
# подключаю pandas и сразу на всякий пожарный подключаю numpy

import pandas as pd
import numpy as np
```

```python
# читаю данные, размещаю их в переменную которая будет датафреймом

df_ratings = pd.read_csv('ml-latest-small/ratings.csv')
df_ratings.head()
```

#### Вариант_1:
Группировка + агрегация несколькоих функций методом "agg"

```python
# группиру данные методом groupby указывая толбик по которому делаю групировку "userId"
# и для последующего вычисления (ltv и фильтрации более 100 оценок) формирую отдельные столбци 'min', 'max', 'count'
# использую метод reset_index, что бы "userId" не переходил в индекс (правда не совсем понимаю требудется ли это, может можно оставить 'usrId' в индексе, не понимаю какие сложности это даст)

grouped_users_ltv = df_ratings.groupby('userId').agg(['min', 'max', 'count'])['timestamp'].reset_index()
grouped_users_ltv.head(2)
```

```python
# формирую отдельный столбик где вычисляю ltv по формуле из условия

grouped_users_ltv['user_ltv'] = grouped_users_ltv['max'] - grouped_users_ltv['min']
grouped_users_ltv.head()
```

```python
# для наглядности выношу отдельно фильтрацию выше 100 оценок

grouped_users_ltv[grouped_users_ltv['count'] > 100].head()
```

#### Вариант_2:
Отдельная Группировка + функция + метод apply
В этом варианте использую сгруппированные данные по 'userId' и расчитанные столбики: 'min', 'max', 'count'

```python
# пишу функцию которая считает ltv d каждой строке

def user_ltv (row):
    return row['max'] - row['min']
```

```python
# беру ранее сгруппированные данные, и отправляю их в функцию, результаты прописываю в отдельный столбик 'user_ltv_func'
# добавляю слово func для понимания что в столбце результаты которые возвращает функция 'user_ltv'
# фильтрую по кол-ву оценок выше 100

grouped_users_ltv['user_ltv_func'] = grouped_users_ltv.apply(user_ltv, axis=1).head()
grouped_users_ltv[grouped_users_ltv['count'] >= 100].head()
```

#### Варинт_3:  
Группировка и одновременное использование функции через apply

```python
# функция считает столбик ltv и только, хотя хотелосьбы что бы еще считала кол-во оценок для последующей фильтрации
# Я не знаю как сделать что бы функция 'func_grouped_user_ltv' одновременно кол-во оценок для последующей фильтрации и считала ltv
# Я считаю что в этом примере расмотрен простой случай, и функция сздесь не нужна, но на перспективу, для более сложных вычислений хочется понять как возвращать функцией несколько результатов в несколько столбцов

def func_grouped_user_ltv (row):
    return row.timestamp.max() - row.timestamp.min()
```

```python
# группирую данные, и одновременно отправляю их для расчета ltv  в функцию

df_ratings.groupby('userId').apply(func_grouped_user_ltv).head()
```

### Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).  


Необходимо сформировать две таблицы:  
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

***Обратите внимание, что в процессе объединения таблиц данные не должны теряться.***

- Дана статистика услуг перевозок клиентов компании по типам:
    - rzd - железнодорожные перевозки
    - auto - автомобильные перевозки
    - air - воздушные перевозки
    - client_base - адреса клиентов

```python
import pandas as pd
import numpy as np
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# заовжу переменную, прописываю df.merge(df c которым хочу объединить, how="тип join")
# в конуе ставлю точку, вызываю merge и повторяю процедуру добоаляя еще таблицу которую хочу объедить в итоговый df

transport_stat = rzd.merge(auto, how='outer').merge(air, how='outer').fillna(0)
transport_stat
```

```python
# добавляю к transport_stat с помощью merge, df c адресами: client_id
# в методе merge предусморенно авто определние столбца "ключа" коим является client_id, но иногда merge не может определить столбци
# здесь я явно указал по каким столбца merge должен объединить df

transport_address_stat = transport_stat.merge(client_base, how='outer', left_on='client_id', right_on='client_id')
transport_address_stat
```
